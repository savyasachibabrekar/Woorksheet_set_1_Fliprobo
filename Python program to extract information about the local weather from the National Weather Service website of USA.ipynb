##### Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.


```python
# Importing the important libraries.

from bs4 import BeautifulSoup
import requests
import re

# get the respone from www.weather.gov  url 
weatherurl = 'https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YPko-egzY2w'
resp_wther = requests.get(weatherurl)
resp_wther
```

```python
#Check the response
resp_wther.content
```

```python
# Using Soup
soup_weather = BeautifulSoup(resp_wther.content)
soup_weather
```

```python
# Declaring the structured data in prettify()
print(soup_weather.prettify())
```

```python
# Extracting the period names from p tags
period = soup_weather.find_all('p', class_='period-name')
period

#<p class="period-name">Overnight<br><br></p>
```

```python
len(period)
```

```python
# Extracting the 7 days periods names.
period_names = []
for i in period:
    period_names.append(i.text)
period_names
```

```python
# Extracting the short description under p tags and class short-decs
shortdesc = soup_weather.find_all('p', class_='short-desc')
shortdesc
# <p class="short-desc" style="height: 67px;">Partly Cloudy</p>
```

```python
len(shortdesc)
```

```python
# Creating the list of short discription 
short_desc = []
for i in shortdesc:
    short_desc.append(i.text)
short_desc
```

```python
# Extracting the temperature under p tags and class = temp temp-low
temp = soup_weather.find_all('p', class_=['temp temp-low','temp temp-high'])
temp
#<p class="temp temp-low">Low: 54 F</p>
# <p class="temp temp-high">High: 66 F</p>
```

```python
len(temp)
```

```python
# Creating the list for temprature 
temprature = []
for i in temp:
    temprature.append(i.text)
temprature
```

```python
# Extracting the discription under p tags and class= col-sm-10 forecast-text
discript = soup_weather.find_all('div', class_='col-sm-10 forecast-text')
discript[:9]
# <div class="col-sm-10 forecast-text">Partly cloudy, with a low around 54. Southwest wind 7 to 9 mph. </div>
```
