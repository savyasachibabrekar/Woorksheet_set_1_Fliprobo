#### 4.Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have toscrape:
i) Top 10 ODI teams in women’s cricket along with the records for matches, points andrating.
ii) Top 10 women’s ODI players along with the records of their team and rating.
iii) Top 10 women’s ODI all-rounder along with the records of their team andrating

```python
# Importing the important libraries.

from bs4 import BeautifulSoup
import requests
import re

# Download IMDB's Top 100 data
icc_url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
resp = requests.get(icc_url)
resp
```

```python
#Check the response
resp.content
```

```python
# Using Soup
soup1 = BeautifulSoup(resp.content)
soup1
```

```python
# Declaring the structured data in prettify()
print(soup1.prettify())
```

```python
# Extracting the team names from tags
teams = soup1.find_all('span', class_='u-hide-phablet')
teams[:10]
```

```python
len(teams)
```

```python
# Extracting top 20 the team names 
team_names = []
for i in teams:
    team_names.append(i.text)
team_names[1:10]
```

```python
matches2 = soup1.find_all('td', class_='table-body__cell u-center-text')
matches2
```

```python
matches_names = []
for i in matches2:
    matches_names.append(i.text)
matches_names[0:18:2]
```

```python
points = soup1.find_all('td', class_='table-body__cell u-center-text')
points[:20]
```

```python
points_names = []
for i in points:
    points_names.append(i.text)
points_names[1:20:2]
```

```python
rating = soup1.find_all('td', class_='table-body__cell u-text-right rating')
rating[:9]
```

```python
rating_names = []
for i in rating:
    rating_names.append(i.text)
rating_names[:10]
```

```python
# Creating the DataFrame in python using the pandas 

import pandas as pd 
women_cricket_df = pd.DataFrame({})
women_cricket_df['Team Names'] = team_names[1:10]
women_cricket_df['Matches'] = matches_names[0:18:2]
women_cricket_df['Points'] = points_names[1:20:2]
women_cricket_df['Ratings'] = rating_names[:10]

women_cricket_df
```

```python
#  Final Top 10 list of Cricket according to their rankings 


top_row = pd.DataFrame({'Team Names':['Australia'],'Matches':[18],'Points':[2955], 'Ratings':[164]})
# Concat with old DataFrame and reset the Index.
women_cricket_df = pd.concat([top_row, women_cricket_df]).reset_index(drop = True)
women_cricket_df
```
