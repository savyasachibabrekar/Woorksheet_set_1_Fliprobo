#### Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
# Importing the important libraries.

from bs4 import BeautifulSoup
import requests
import re

# get the respone from the url 'www.amazon.com' 
amzn_url = 'https://www.amazon.in/s?k=mobile+under+20000&i=electronics&rh=n%3A976419031%2Cp_36%3A-2000000&dc&crid=33F9Y9CL4VNYG&qid=1626864687&rnid=1318502031&sprefix=mobile+%2Caps%2C439&ref=sr_nr_p_36_1'
amzn_resp = requests.get(amzn_url)
amzn_resp
```

```python
#Check the response
amzn_resp.content
```

```python
# Using Soup
soup_amazon = BeautifulSoup(amzn_resp.content)
soup_amazon
```

```python
# Declaring the structured data in prettify()
print(soup_amazon.prettify())
```

```python
# Extracting the product names from tags
product = soup_amazon.find_all('span', class_='a-size-medium a-color-base a-text-normal')
product
# <span class="a-size-medium a-color-base a-text-normal">Samsung Galaxy M31s (Mirage Black, 6GB RAM, 128GB Storage) 6 Months Free Screen Replacement for Prime</span>
```

```python
# Final Product name list in the dataframe 

productname = []
for i in product:
    productname.append(i.text)
productname
```

```python
len(productname)
```

```python
price = soup_amazon.find_all('span', class_='a-price-whole')
price
```

```python
# Final Product price list in the dataframe 

productprice = []
for i in price:
    productprice.append(i.text)
productprice
```

```python
len(productprice)
```

```python
# extracting the Image Url for the <a href > tags
img_url = soup_amazon.find_all('img', class_='s-image')
img_url

#img_url = soup_amazon.find_all('a', class_='a-link-normal s-no-outline', href = True)
#img_url
```

```python
image_url = []
for i in img_url:
    image_url.append(i.get('src'))
image_url
```

```python
len(image_url)
```

```python
avg_rating = soup_amazon.find_all('span', class_='a-icon-alt')
avg_rating[:22]

# <span class="a-icon-alt">4.3 out of 5 stars</span>
```

```python
len(avg_rating)
```

```python

average_rating = []
for i in avg_rating:
     average_rating.append(i.text)
average_rating
```

```python
# Final List Includes Product Name, Product Price, Image URL, Average Ratings


import pandas as pd 
amazon_df = pd.DataFrame({})
amazon_df['Product Names'] = productname[:22]
amazon_df['Product Price'] = productprice[:22]
amazon_df['URL'] = image_url[:22]
amazon_df['Ratings'] = average_rating[:22]

amazon_df
```
